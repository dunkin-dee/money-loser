```python
def get_sma(candles, fac=5):
    simple_moving_averages = {}
    times = sorted(list(candles.index))   
    for list_x in range(fac-1, len(times)):
        average_sum = 0
        working_times = times[list_x-fac+1:list_x+1]
        for time in working_times:
            average_sum += candles.loc[time]["close"]
        simple_average = average_sum/fac
        simple_moving_averages[working_times[-1]] = simple_average
    return(simple_moving_averages)
```

```python
def get_sma2(df, fac=5):
    averages = pd.Series(df["close"])
    for loop_iter in range(1, fac):
        averages = averages + df.shift(loop_iter)["close"]
    averages = averages/fac
    averages.dropna()
    return(averages[fac-1:])
```

```python

def get_single_candlestick_patterns(candles):
    patterns = {
        "hammers": [],
        "ihammers": [],
    }

    times = sorted(list(candles.index))
    mean_body_size =  candles['bodysize'].mean()



    for time in times:
        curr_row=candles.loc[time]

        #solving for hammers
        hammer_valid_size = curr_row["bodysize"] < (mean_body_size/1.5)
        if curr_row["direction"] == "up" :
            hammer_valid_wick = abs(curr_row["high"] - curr_row["close"]) < (curr_row["bodysize"]*1.5)
            hammer_valid_shadow = abs(curr_row["open"] - curr_row["low"]) > (curr_row["bodysize"]*2)
        else:
            hammer_valid_wick = abs(curr_row["high"] - curr_row["open"]) < (curr_row["bodysize"]*1.5)
            hammer_valid_shadow = abs(curr_row["close"] - curr_row["low"]) > (curr_row["bodysize"]*2)

        if hammer_valid_size and hammer_valid_shadow and hammer_valid_wick:
            patterns["hammers"].append(time)

        #solving for inverted hammers

        if curr_row["direction"] == "up" :
            ihammer_valid_wick = abs(curr_row["high"] - curr_row["close"]) > (curr_row["bodysize"]*3)
            ihammer_valid_shadow = abs(curr_row["open"] - curr_row["low"]) < (curr_row["bodysize"]*1.5)
        else:
            ihammer_valid_wick = abs(curr_row["high"] - curr_row["open"]) > (curr_row["bodysize"]*3)
            ihammer_valid_shadow = abs(curr_row["close"] - curr_row["low"]) < (curr_row["bodysize"]*1.5)

        if hammer_valid_size and ihammer_valid_shadow and ihammer_valid_wick:
            patterns["ihammers"].append(time)



    return(patterns)
```

```python
def get_single_candlestick_patterns(df):
    """
    accepts dataframe with bodysize, wick, shadow and returns dict
    """
    df2 = df.copy(deep=True)
    mbs = df['bodysize'].mean()
    
    df2.loc[(df2["wick"]<df2["bodysize"]*1.5) & (df2["shadow"]>df2["bodysize"]*3) & (df2["bodysize"]<mbs/2), "hammers"] = 1
    df2.loc[(df2["wick"]>df2["bodysize"]*3) & (df2["shadow"]<df2["bodysize"]*1.5) & (df2["bodysize"]<mbs/2), "ihammers"] = 1

    patterns = {}
    patterns["hammers"] = list(df2[df2["hammers"]==1].index)
    patterns["ihammers"] = list(df2[df2["ihammers"]==1].index)
    return(patterns)
```

```python
def get_double_candlestick_patterns(candles):
    bullish_engulfing = []
    bearish_engulfing =[]
    tweezer_bottoms = []
    tweezer_tops = []

    
    times = sorted(list(candles.index))
    mean_body_size =  candles['bodysize'].mean()

    for time in range(1, len(times)):
        curr_row = candles.loc[times[time]]
        prev_row = candles.loc[times[time-1]]

        #bullish engulfing
        bu_valid_directions = curr_row["direction"] == "up" and prev_row["direction"] == "down"
        valid_sizes = curr_row["bodysize"] > prev_row["bodysize"]

        if bu_valid_directions and valid_sizes:
            bullish_engulfing.append(times[time])

        #bearish engulfing
        be_valid_directions = curr_row["direction"] == "down" and prev_row["direction"] == "up"
        if be_valid_directions and valid_sizes:
            bearish_engulfing.append(times[time])

        #tweezer bottoms
        tb_valid_directions = curr_row["direction"] == "up" and prev_row["direction"] == "down"
        tb_valid_tops = abs(curr_row["high"] - prev_row["high"]) < (mean_body_size/4)
        curr_shadow = abs(curr_row["open"] - curr_row["low"])
        prev_shadow = abs(prev_row["close"] - prev_row["low"])
        tb_valid_shadow = abs(prev_shadow - curr_shadow) < (mean_body_size/4) and curr_shadow > mean_body_size

        if tb_valid_directions and tb_valid_tops and tb_valid_shadow:
            tweezer_bottoms.append(times[time])

        #tweezer tops
        tt_valid_directions = curr_row["direction"] == "down" and prev_row["direction"] == "up"
        tt_valid_tops = abs(curr_row["low"] - prev_row["low"]) < (mean_body_size/4)
        curr_wick = abs(curr_row["high"] - curr_row["open"])
        prev_wick = abs(prev_row["high"] - prev_row["close"])
        tt_valid_wick = abs(prev_wick - curr_wick) < (mean_body_size/4) and curr_wick > mean_body_size

        if tt_valid_directions and tt_valid_tops and tt_valid_wick:
            tweezer_tops.append(times[time])
        
    
    bullish_patterns = {
        "bullishengulfing": bullish_engulfing,
        "tweezerbottoms": tweezer_bottoms,
    }

    bearish_patterns ={
        "bearishengulfing": bearish_engulfing,
        "tweezertops": tweezer_tops,
    }

    double_candlestick_patterns = {
        "bullish": bullish_patterns,
        "bearish": bearish_patterns,
    }

    return(double_candlestick_patterns)
```

```python
def get_double_candlestick_patterns_2(df):
    """
    Accepts dataframe with bodysize, direction, wick, shadow and returns dict
    """
    mbs = df["bodysize"].mean()
    df2 = df.copy(deep=True)
    df3 = df2.shift(1)
    for col_name in df3.columns:
        df3.rename({col_name:"prev"+col_name},axis='columns',inplace=True)
    df2 = pd.concat([df2, df3], axis=1)
    
    df2.loc[(df2["direction"]=="up")&(df2["prevdirection"]=="down")&(df2["bodysize"]>df2["prevbodysize"]),"bullish_engulfing"] = 1
    df2.loc[(df2["direction"]=="down")&(df2["prevdirection"]=="up")&(df2["bodysize"]>df2["prevbodysize"]),"bearish_engulfing"] = 1
    df2.loc[(df2["wick"]>mbs)&(df2["prevwick"]>mbs)&(df["shadow"]<mbs/4),'tweezer_tops'] = 1
    df2.loc[(df2["shadow"]>mbs)&(df2["prevshadow"]>mbs)&(df["wick"]<mbs/4),'tweezer_bottoms'] = 1
    
    bullish={}
    bearish={}
    
    bullish["bullish_engulfing"] = list(df2[df2["bullish_engulfing"]==1].index)
    bullish["tweezer_bottoms"] = list(df2[df2["tweezer_bottoms"]==1].index)
    bearish["bearish_engulfing"] = list(df2[df2["bearish_engulfing"]==1].index)
    bearish["tweezer_tops"] = list(df2[df2["tweezer_tops"]==1].index)
    return({"bullish":bullish, "bearish":bearish})
```
